```python
# Import libbraries
import requests
import bs4 # Beautiful Soup to parse html
import os
import glob
import shutil
```

```python

def jid2name(jid):
### The function transfer jid.html to a name-of-the-journals
### input: jid
### output: name-of-the-journals
    res = requests.get('http://www.natscipub.org/Home/Journals/journal_articles/jid/' + str(jid))
    res.raise_for_status() # flag check for any status error
    res.status_code # 200 means normal
    soup = bs4.BeautifulSoup(res.text, 'html.parser')
    title = soup.select('#wrapwidthJ > div > div.journals-title > p')
    title1 = title[0].text.lower()
    ### string manuipluation
    # Cleaning format to generate new title
    title2 = title1.replace('american','')
    title2 = ''.join([ch for ch in title2 if ch.isalpha() or ch ==' '])
    title2 = title2.split()
    titleNew = '-'.join(title2)
    return titleNew
```
